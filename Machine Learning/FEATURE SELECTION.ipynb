```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris = load_iris()
```

```python
X = iris.data
```

```python
X
```

```python
y = iris.target
```

```python
# we going to remove low variance feature
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=0.5)
```

```python
X_vt = sel.fit_transform(X)
```

```python
X_vt
```

```python
fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(X_vt[:,0],X_vt[:,1],X_vt[:,2],c=y,cmap='rainbow',s=50)
plt.show()
```

```python
sel1 = VarianceThreshold(threshold=0.6)
```

```python
X_vt1 = sel1.fit_transform(X)
```

```python
X_vt1
```

```python
plt.scatter(X_vt1[:,0],X_vt1[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
# Select from Model
```

```python
from sklearn.svm import SVC
from sklearn.feature_selection import SelectFromModel
```

```python
model = SVC(kernel='linear')
```

```python
model.fit(X,y)
```

```python
sfm = SelectFromModel(model,prefit=True)
```

```python
X_sfm = sfm.transform(X)
```

```python
X_sfm
```

```python
plt.scatter(X_sfm[:,0],X_sfm[:,1],c=y,cmap='rainbow')
plt.show()
```
